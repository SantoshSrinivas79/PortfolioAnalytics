```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import quandl as ql
```

```python
# Data Query and Range Delimiting
cashquery = ql.get("USTREASURY/YIELD.1", collapse="monthly")
mcash = cashquery['2007-09-01':'2017-09-30']
```

```python
cashquery
```

```python
mcash
```

```python
# Monthly and Annual Returns Calculation
mcashret = (mcash / 100) / 12
mcashret.columns = ['mcashret']
acashret = mcash / 100
acashret.columns = ['acashret']
```

```python
mcashret
```

```python
acashret
```

```python
# Cumulative Monthly Returns Calculation
mcashcum = np.cumprod(mcashret + 1) - 1
mcashcum.columns = ['mcashcum']
```

```python
# Annualized Monthly Returns Calculation
mcashcuma = np.cumprod(mcashret + 1) ** (12 / len(mcashret)) - 1
mcashcuma.columns = ['mcashcuma']
# Obtain last Annualized Monthly Return
mcashretailoc = mcashcuma.iloc[len(mcashcuma) - 1]
mcashreta = mcashretailoc.values
```

```python
# Annualized Monthly Standard Deviation Calculation
mcashstdeva = np.std(mcashret) * np.sqrt(12)
```

```python
# Monthly Returns Chart
mcashret.plot()
plt.title('1 Month U.S. Treasury Bills Monthly Returns')
plt.legend(loc='upper right')
plt.show()
```

```python
# Cumulative Monthly Returns Chart
mcashcum.plot()
plt.title('1 Month U.S. Treasury Bills Cumulative Monthly Returns')
plt.legend(loc='upper left')
plt.show()
```

```python
# Annualized Monthly Return and Standard Deviation Table
cashdata = [{'0': '', '1': 'mcashret'},
        {'0': 'Annualized Return', '1': mcashreta},
        {'0': 'Annualized Standard Deviation', '1': mcashstdeva.values}]
cashtable = pd.DataFrame(cashdata)
print("")
print(cashtable)
print("")
#Cash has low performance with low risk
```
